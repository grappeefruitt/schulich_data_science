```python
data = data_13_14.copy() # make a copy to shorten data set name
```

```python
#isolate the column 'hour'
data['hour'] = data.index.map(lambda x: x[11:])
```

```python
grouped_data = data.groupby('hour')
average_curves = grouped_data.agg('mean')
average_curves.shape
```

```python
avg_curves_norm = average_curves/average_curves.mean()
```

```python
avg_curves_norm
```

```python
avg_curves_norm[['MT_001','MT_002','MT_369','MT_370']].plot()
```

```python
#keep an original copy of the dataset using copy
#put data in arrary format when doing machine leanrng
X = avg_curves_norm.copy()
X = np.array(X.T)
```

```python
X
#set up data for approimate machine leanring 
```

```python
del silhouette_score
```

```python
inertia = []
silhouette_avg = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    labels = kmeans.labels_
    centers = kmeans.cluster_centers_
    if i>1:
        silhouette_avg = silhouette_avg + [silhouette_score(X, labels)]

# Silhouette Score

plt.plot(range(1, 15), inertia)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()


plt.plot(range(2, 15), silhouette_avg)
plt.title('Silhouette Score')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()

#pick 6/8 for the first graoh and 4/6 for sil score
#we want low inter but high sil score
```

```python
K = 6
kmeans = KMeans(n_clusters=K, random_state=0).fit(X)
```

```python
# create plots

for c in range(K):
    avg_curves_norm.T[kmeans.labels_==c].T.plot(legend=False)
    plt.plot(kmeans.cluster_centers_[c], linewidth=8, color='orange')
    plt.title("cluster "+str(c))
```

```python
X = []
for J in range(2*365):
    X.extend([np.array(oneclient[J*96:(J+1)*96])])
```

```python
X
```

```python
from datetime import date, timedelta

d1 = date(2013,1,1)
d2 = date(2014, 12, 31)
delta = d2-d1
daysyear=[]
D = {0:'mon', 1:'tue', 2:'wed', 3:'thu', 4:'fri',5:'sat', 6:'sun'}
for i in range(delta.days + 1):
    daysyear.extend([D[(d1+timedelta(days=i)).weekday()]+"-"+str(d1+timedelta(days=i))])
```

```python
daysyear
```

```python
inertia_1 = []
silhouette_avg_1 = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    inertia_1.append(kmeans.inertia_)
    labels = kmeans.labels_
    centers = kmeans.cluster_centers_
    if i>1:
        silhouette_avg_1 = silhouette_avg_1 + [silhouette_score(X, labels)]

# Silhouette Score

plt.plot(range(1, 15), inertia_1)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()


plt.plot(range(2, 15), silhouette_avg_1)
plt.title('Silhouette Score')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()

#the optimal number is 2
#as it bends on 2 on the first graph and is the hghest on 2 on the seond graph
```
