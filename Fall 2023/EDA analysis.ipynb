```python
df.shape
```

```python
#converting object dtype to datatime format
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df = df.rename(columns={'coaster_name':'Coaster_Name','year_introduced':'Year_Introduced','opening_date_clean':'Opening_Date','speed_mph':'Speed_mph','height_ft':'Height_ft','Inversions_clean':'Inversions',
'Gforce_clean':'Gforce'})
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])] \
    .reset_index(drop=True).copy()
```

### Visualization to explore relationships

```python
df.plot(kind='scatter',
        x='Speed_mph',
        y='Height_ft',
        title='Coaster Speed vs. Height')
plt.show()
```

```python
ax = sns.scatterplot(x='Speed_mph',
                y='Height_ft',
                hue='Year_Introduced',
                data=df)
ax.set_title('Coaster Speed vs. Height')
plt.show()
```

```python
sns.pairplot(df,
             vars=['Year_Introduced','Speed_mph',
                   'Height_ft','Inversions','Gforce'],
            hue='Type_Main')
plt.show()
```

```python
df_corr = df[['Year_Introduced','Speed_mph',
    'Height_ft','Inversions','Gforce']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
```
