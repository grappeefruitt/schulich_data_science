### Dataset:
This dataset describes the moose  hunting activity and harvests from year 2006 to 2022.This data breaks down bull moose, cow moose, calf moose and total harvest numbers by wildlife management unit (WMU) and calendar year. Harvest and active hunter numbers are estimates based on replies received from a sample of hunters and are therefore subject to statistical error.

```python

resource_id = '61d62c0e-784a-4061-8bbe-8ee8a42afde7'
limit = 500
offset = 0

api_url = 'https://data.ontario.ca/api/3/action/datastore_search?resource_id={}&limit={}&offset={}'.format(resource_id,limit,offset)

print(api_url)
```

```python
Moose_hunting= pd.DataFrame(data['result']['records'])
Moose_hunting
```

```python
Moose_hunting.dropna()
```

```python
dup = Moose_hunting[Moose_hunting.duplicated()]
dup
#No duplicates are found.
```

```python
#Calculate Total Harvest per Year
total_harvest = Moose_hunting.groupby(['Year'])['Total Harvest'].sum().reset_index()
```

```python
#Calculate Average Harvest per Year
total_harvest['Average Harvest'] = total_harvest['Total Harvest'] / Moose_hunting['Active Hunters']
```

```python
#Merge the Average Harvest column back into the original DataFrame
Moose_hunting = Moose_hunting.merge(total_harvest[['Year', 'Average Harvest']], on=['Year'], how='left')
```
